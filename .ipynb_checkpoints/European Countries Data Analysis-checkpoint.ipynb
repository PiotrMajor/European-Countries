```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

#read csv files
pollution = pd.read_csv('pollution_2016.csv')
life_satisfaction = pd.read_csv('life_satisfaction_2013.csv')
income = pd.read_csv('median_income_2016.csv')
job_satisfaction = pd.read_csv('job_satisfaction_2013.csv')
leisure_satisfaction = pd.read_csv('leisure_satisfaction_2013.csv')
weather = pd.read_csv('weather.csv')
life_expectancy = pd.read_csv('life_expectancy_2016.csv')
unemployment = pd.read_csv('unemployment_2016.csv')
gdp = pd.read_csv('gdp_2016.csv')
health = pd.read_csv('perceived_health_2016.csv')
population = pd.read_csv('population_2011.csv')
work_hours = pd.read_csv('work_hours_2016.csv')

# merge all the datasets on the country column.

x = pd.merge(pollution, life_satisfaction, on = 'country')
x1 = pd.merge(x, job_satisfaction, on = 'country')
x2 = pd.merge(x1, life_satisfaction, on = 'country')
x3 = pd.merge(x2, income, on = 'country')
x4 = pd.merge(x3, leisure_satisfaction, on = 'country')
x5 = pd.merge(x4, weather, on = 'country')
x6 = pd.merge(x5, life_expectancy, on = 'country')
x7 = pd.merge(x6, unemployment, on = 'country')
x8 = pd.merge(x7, gdp, on = 'country')
x9 = pd.merge(x8, health, on = 'country')
x10 = pd.merge(x9, population, on = 'country')
df = pd.merge(x10, work_hours, on = 'country')

# drop some of the useless columns present in the dataset
df = df.drop(['Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9',
              'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13', 'Unnamed: 14',
              'Unnamed: 15', 'Unnamed: 16','prct_life_satis_high_y', 'prct_life_satis_med_y',
              'prct_life_satis_med_x', 'prct_life_satis_low_x', 'prct_life_satis_low_y',
              'avg_high_temp', 'avg_low_temp', 'prct_yng_adt_pop', 'prct_job_satis_med',
              'prct_job_satis_low', 'prct_leisure_satis_med', 'prct_leisure_satis_low', 'prct_health_good', 'prct_health_fair', 'prct_health_bad', 'prct_health_verybad'], axis = 1)

df
```

Let's see what European countries have the highest life satisfaction

```python
# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(6, 10))

# Plot the high life satisfaction
sns.set_color_codes("muted")
sns.barplot(x="prct_life_satis_high_x", y="country", data=df.sort_values(by=['prct_life_satis_high_x'], ascending=False), color="b")

# Add a legend and informative axis label
ax.set(xlim=(0, 45), ylabel="", 
       xlabel="Percentage of people with high life satisfaction");
```

Based on the bar chart, it looks like citizens of Denmark have the highest life satisfaction, followed by Finland and Switzerland. The least satisfied people are in Bulgaria, Turkey, and Hungary. Let's see what correlates to the happines.

```python
# initialize the correlation matrix
corrMatrix = df.corr()

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

fig, ax = plt.subplots(figsize=(5,7))
sns.heatmap(corrMatrix[['prct_life_satis_high_x']].sort_values(by=['prct_life_satis_high_x'],ascending=False), annot=True, cmap=cmap, vmin=-1);
```

The correlation matrix shows that job, leisure satisfaction, and income are positively correlated to life satisfaction. On the other hand, the more hours people spend working, the less likely they are to be satisfied with their life.

Next, let's analyze income in different countries

```python
# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(6, 10))

# Plot annual income
sns.set_color_codes("muted")
sns.barplot(x="median_income", y="country", data=df.sort_values(by=['median_income'], ascending=False), color="b")

# Add a legend and informative axis label
ax.set( ylabel="",
       xlabel="Median annual income per capita in US $");
```

Luxembourg has the highest median annual income, followed by Switzerland and Norway. The poorest people live in Romania, Turkey and Bulgaria.

The next chart will look into how much time people spend working.

```python
# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(6, 10))

# Plot hours worked per week
sns.set_color_codes("muted")
sns.barplot(x="avg_hrs_worked", y="country", data=df.sort_values(by=['avg_hrs_worked'], ascending=False), color="b")

# Add a legend and informative axis label
ax.set(xlim=(0, 50), ylabel="",
       xlabel="Average hours worked per week");
```

Turkey have the longest working hours (46.8 per week), while the citizens of Netherlands work only 30.3 hours per week.

Next, I'm creating a new column to estiamte average income per hour spent at work.

```python
#Create a new column hours median income / hours worked per week
df['median_income_per_hour'] = df['median_income'] / (df['avg_hrs_worked']*52)
df
```

Plotting the results from the new column.

```python
# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(6, 10))

# Plot the income / hour worked
sns.set_color_codes("muted")
sns.barplot(x="median_income_per_hour", y="country", data=df.sort_values(by=['median_income_per_hour'], ascending=False), color="b")

# Add a legend and informative axis label
ax.set( ylabel="",
       xlabel="Median income per hour");
```

Countries with the highest median salary per hour include

* Norway - \$15.7 per hour
* Switzerland - \$15.3 per hour
* Luxembourg -\$14.7 per hour

Countries with the lowest median salary include

* Romania - \$2.3 per hour
* Turkey - \$2.7 per hour
* Bulgaria - \$3.2 per hour

There are many factors that correlate to Europeans’ life satisfaction. One of the most important elements is work-life balance and annual income.

It appears that Scandinavian countries (Denmark, Norway, Iceland, Finland, Sweden), as well as Switzerland, are the best countries to live in.
